```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
rng = np.random.default_rng(2021)

z = rng.multivariate_normal([0,0], cov=[ [1, 0], [0,1 ]], size=10000)
```

```python
plt.scatter(z[:,0], z[:,1], s=1, alpha=.2);
plt.gca().set_aspect(True)
```

```python
def xfrom(z):
    g = z/5 + z / np.linalg.norm(z, axis=1).reshape(-1, 1)  # (100,) -> (100, 1) for broadcasting
    return g
```

```python
g = xfrom(z)

circle = plt.Circle((0,0), radius=1, fill=False, alpha=.5, color='r')
ax = plt.gca()
ax.add_patch(circle);

plt.scatter(g[:,0], g[:,1], s=1, alpha=.2);
plt.gca().set_aspect(True)
```

```python
flag =  (-0.1 < g[:,1]) & (g[:,1] < 0.1)
```

```python
(-.1 < g[:,1] )  &  (g[:,1] < .1)
```

```python
a = g[flag,:]
a.shape
```

```python
plt.hist(a[:,0], bins=50, edgecolor='r');
plt.title('density along the line $y=0$');
```

```python
sns.displot(a[:,0], bins=50);
```
