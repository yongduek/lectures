```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)
f = y * x * x 
```

```python
from matplotlib import cm 

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(x, y, f, cmap=cm.coolwarm)
fig.colorbar(surf, shrink=.5)
```

```python
plt.figure(figsize=(10,10))
plt.contourf(x, y, f, levels=200, cmap=cm.RdBu_r)
plt.colorbar()
circle = plt.Circle( (0,0), 1, color='k', fill=False)
ax = plt.gca()
ax.add_patch(circle)
ax.set_aspect('equal')
```

```python
import numpy as np 

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)
f = y * y -  x  # y^2 - x
```

```python
import matplotlib.pyplot as plt 
from matplotlib import cm 

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(x, y, f, cmap=cm.coolwarm)
fig.colorbar(surf, shrink=.5);
```

```python
plt.figure(figsize=(7, 7))
plt.contourf(x, y, f, levels=200, cmap=cm.RdBu_r)
plt.colorbar()
circle = plt.Circle( (0,0), 1, color='k', fill=False)
ax = plt.gca()
ax.add_patch(circle)
ax.set_aspect('equal')
```
