```python
import numpy as np 
import matplotlib.pyplot as plt 
```

$ \mu_t = \beta \mu_{t-1} + (1-\beta) y_t $

```python
beta = 0.98
N = 200
omb = 1 - beta 
weights = np.array([omb * beta**k for k in range(N)])
flags = weights / weights[0] > (1/np.e)
print(flags.sum())
print(weights.sum() + beta**N)
plt.plot(weights);
plt.plot(flags*.12, '-o')
plt.grid(True)
```

```python
for beta in [0.9, 0.92, 0.95, 0.97, 0.98]:
    omb = 1 - beta 
    weights = np.array([omb * beta**k for k in range(200)])
    plt.plot(weights / weights[0]);
```
