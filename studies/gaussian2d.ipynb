```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as stats 
```

```python
# define a 2D Gaussian
cov = np.array( [[1, 0.8], [0.8, 1]] )
G = stats.multivariate_normal(mean=[0,0], cov=cov)
```

```python
# make 2D grid for plotting with plt
gridstep=0.1
x, y = np.mgrid[-5:5:gridstep, -3:3:gridstep]
pos = np.dstack((x,y))
```

```python
plt.contourf(x,y, G.pdf(pos));
plt.gca().set_aspect('equal');
```

```python
# eigen decomposition of Cov matrix
cov = np.array( [[1, 0.8], [0.8, 1]] )
np.linalg.eig(cov)
```

```python
import superimport

from scipy.stats import binom

import numpy as np
import matplotlib.pyplot as plt
# import pyprobml_utils as pml



N = 10

thetas = [.25, .5, .75, .9]

X = np.arange(0, N + 1)

for th in thetas:
    probs = binom.pmf(X, N, th)
    title = r'$\theta$' + "=" + str(th)
    fig, ax = plt.subplots()
    ax.bar(X, probs, align='center')
    plt.xlim([min(X)-.5, max(X)+.5])
    plt.xticks(X)
    plt.title(title)
    plt.draw()
    # pml.savefig('binomDistTheta' + str(int(th*100)) + '.pdf')

plt.show()
```
