If the policy is $\pi(a) = [1-p, p]$, the state-value function $v_\pi(s)$ depends on the probability $p$. 

Here, we want to compute $v_\pi(s=0, p)$ as a function of $p$, where $s=0$ is the left-most grid.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def v(p):
    A = np.array([ [(1-p), p, 0],
                  [p, 0, 1-p],
                  [0, 1-p, 0]
                 ])
    B = np.eye(3) - A
    b = np.array([-1,-1,-1])
    
    v = np.linalg.inv(B) @ b
    return v
```

```python
v(.5)
```

```python
prange = np.linspace(0.01, 0.9999, 10000)
v0s = [v(p)[0] for p in prange]
plt.plot(prange, v0s)
plt.ylim((-100, 0));
xmax = prange[np.argmax(v0s)]
ymax = np.max(v0s)
plt.text(x=xmax, y=ymax+3, s=f'max={ymax:.2f}@{xmax:.2f}');
plt.scatter(x=xmax, y=ymax);
plt.xlabel('Policy: probability of choosing [right] at each state');
plt.ylabel('$V(0)$ with the policy.');
```
