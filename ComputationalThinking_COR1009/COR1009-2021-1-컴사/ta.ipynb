```python
def sigmoid(x):
    return 1./(1 + np.exp(-x))
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10, 10)

# ax.plot(x, 1./(1 + np.exp(-x)))
ax.plot(x, sigmoid(x+.5), color=(1, 0, 0))
ax.plot(x, sigmoid(x+1.5), color=(0, 1, 0))
ax.plot(x, sigmoid(x+2.5), color=(0, 0, 1))
ax.plot(x, sigmoid(x+3.5), color='#ff0000')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10, 10)

# ax.plot(x, 1./(1 + np.exp(-x)))
ax.plot(x, sigmoid(x+3.5), color='#ff0000')
ax.plot(x, sigmoid(x+3.5), color='#00aa00')
```
